# Convex hull of polygons

```julia
using Plots
using Revise
```

```julia
using PolygonAlgorithms
using PolygonAlgorithms: x_coords, y_coords
```

```julia
poly = [
        (2.0, 10.0), (4.0, 10.0), (4.0, 6.0), (6.0, 6.0), (6.0, 10.0), (8.0, 10.0), 
        (8.0, 1.0), (6.0, 1.0), (6.0, 4.0), (4.0, 4.0), (4.0, 1.0), (2.0, 1.0)
    ]
;
```

```julia
idxs = vcat(1:length(poly), 1)
plot(x_coords(poly[idxs]), y_coords(poly[idxs]), aspectratio=:equal, marker=:circle)
```

```julia
hull = convex_hull(poly, PolygonAlgorithms.GrahamScanAlg());
```

```julia
hull = vcat(hull, hull[1])
plot!(x_coords(poly[hull]), y_coords(poly[hull]))
scatter!(x_coords(poly[hull]), y_coords(poly[hull]))
```

## Random points

```julia
points = [
        (0.5, 0.7),
        (0.3, 0.4),
        (0.4, 0.4),
        (0.6, 1.0),
        (0.6, 0.4),
        (0.5, 0.9),
        (0.4, 0.3),
        (0.0, 0.3),
        (0.8, 0.1),
        (0.3, 0.9),
        (0.8, 0.8),
        (0.9, 0.2),
        (0.5, 0.2),
        (0.7, 0.6),
    ];
```

```julia
@time hull = convex_hull(points, PolygonAlgorithms.GiftWrappingAlg());
```

```julia
hull = vcat(hull, hull[1])
scatter(x_coords(points), y_coords(points))
plot!(x_coords(points[hull]), y_coords(points[hull]))
scatter!(x_coords(points[hull]), y_coords(points[hull]))
```

```julia
@time hull1 = convex_hull(points, PolygonAlgorithms.GiftWrappingAlg());
@time hull2 = convex_hull(points, PolygonAlgorithms.GrahamScanAlg());
issetequal(hull1, hull2)
```
